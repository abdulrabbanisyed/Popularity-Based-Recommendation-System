## Popularity Based Recommendation System

```python
import pandas as pd
```

### Movie and Ratings Dataset

```python
Movie_data = pd.read_csv("Your Data path location")
Movie_data
```

```python
Ratings_Data = pd.read_csv("/your data path location")
Ratings_Data
```

### Merging both the data sets into one using "MERGE" function of pandas

```python
Merged_data = pd.merge(Movie_data, Ratings_Data, on= 'movieId')
Merged_data
```

### Criteria for Popularity based Recommendation system is:
###     1. Movie ratings
###     2. Number of views

### Average rating of the movie with respect to title

```python
Merged_data.groupby('title')['rating'].mean().sort_values(ascending= False).head(10)
```

```python
Merged_data.groupby('title')['rating'].count().sort_values(ascending= False).head(10)
```

### Average ratings of the movie against number of views

```python
Average_ratings = pd.DataFrame(Merged_data.groupby('title')['rating'].mean())
Average_ratings
```

```python
Average_ratings['Number of views'] = pd.DataFrame(Merged_data.groupby('title')['rating'].count())
Average_ratings
```

```python
Average_ratings['rating'] = round(Average_ratings['rating'], 1)
Average_ratings
```

### Top 10 movies recommended based on popularity

```python
Top_Movies_recommended = Average_ratings[(Average_ratings['rating'] > 4) & (Average_ratings['Number of views'] > 1000)]
Top_Movies_recommended.head(10)
```

### Top 10 movies recommended based on rating

```python
Top_Movies_recommended_by_rating = Top_Movies_recommended.sort_values(by = 'rating', ascending= False)
Top_Movies_recommended_by_rating.head(10)
```

### Top 10 movies recommended based on number of views

```python
Top_Movies_recommended_by_Number_of_Views = Top_Movies_recommended.sort_values(by = 'Number of views', ascending= False)
Top_Movies_recommended_by_Number_of_Views.head(10)
```
